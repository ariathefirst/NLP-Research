```python
def read_input(input_file):
    """This method reads the input file which is in gzip format"""
    
    logging.info("reading file {0}...this may take a while".format(input_file))
    
    with gzip.open (input_file, 'rb') as f:
        for i, line in enumerate (f): 

            if (i%10000==0):
                logging.info ("read {0} reviews".format (i))
            # do some pre-processing and return a list of words for each review text
            yield gensim.utils.simple_preprocess (line)

# read the tokenized reviews into a list
# each review item becomes a serries of words
# so this becomes a list of lists
documents = list (read_input (data_file))
logging.info ("Done reading data file")
```

```python
#Analyze effect of different Epochs
model_10 = gensim.models.Word2Vec (documents, size=200, window=10, min_count=2, workers=15)
model_10.train(documents,total_examples=len(documents),epochs=20)

model_20 = gensim.models.Word2Vec (documents, size=200, window=20, min_count=2, workers=15)
model_20.train(documents,total_examples=len(documents),epochs=20)

model_30 = gensim.models.Word2Vec (documents, size=200, window=30, min_count=2, workers=15)
model_30.train(documents,total_examples=len(documents),epochs=20)

model_40 = gensim.models.Word2Vec (documents, size=200, window=40, min_count=2, workers=15)
model_40.train(documents,total_examples=len(documents),epochs=20)
```

```python
#get most similar words in model
def analyze_similarity_vector(word, model, num):
    return model.wv.most_similar (positive=word, topn = num)

#top 10 words similar to polite
w1 = ["polite"]
print("Model 10: " + str(analyze_similarity_vector(w1, model_10, 10)))
print("Model 20: " + str(analyze_similarity_vector(w1, model_20, 10)))
print("Model 30: " + str(analyze_similarity_vector(w1, model_30, 10)))
print("Model 40: " + str(analyze_similarity_vector(w1, model_40, 10)))
```
